# 1:matplotlibの概要

## matplotlibとは
* 2次元グラフを描画するためのライブラリ
* 可視化に使うよ
* スタイルは2種類あるｋど、この本はオブジェクト指向スタイルを使うよ

```python
import matplotlib.pyplot as plt
import matplotlib.style
import japanize_matplotlib

matplotlib.style.use('ggplot')
```

## MATLABスタイル
* 数値解析ソフトのMATLABと似た形式で描写する

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

plt.plot(x,y) # 折れ線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定

plt.show() # グラフを表示
```

## オブジェクト指向スタイル
* **描画オブジェクト**に対して**サブプロット**を追加して、サブプロットに対してグラフを描画する
* 複数のグラフをまとめて表示できるよ

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x,y) # 折れ線グラフを描画
ax.set_title('OOP-style') # タイトルを設定

plt.show()
```

# 2:描画オブジェクト

## 描画オブジェクトとサブプロット
* 描画オブジェクト:figureを生成して、その中に1つ以上のサブプロット：subplotを配置する
* 1の例だとfigが描画オブジェクト、その中のサブプロットがax
* 複数サブプロットを配置してみるとこうなる

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2) #サブプロットを2つ配置
plt.show()
```

```python
fig, axes = plt.subplots(2,2) # 2行2列のサブプロットを配置
plt.show()
```

* キーワード引数nrows,ncolsで行、列の数を指定することもできる
* デフォルト値は1だから、列数だけでも成立する

```python
fig, axes = plt.subplots(ncols=2) # 1行2列のサブプロットを配置
plt.show()
```

## グラフのスタイル
* 線の色、太さ、背景色とかを指定できるよ
* 26種類くらい容易されてるよ

```python
import matplotlib.style

print(matplotlib.style.available) # スタイルの一覧を表示
```

```python
# グラフのスタイルにclassicを指定
matplotlib.style.use('ggplot')

fig,ax = plt.subplots()
ax.plot([1,2])
ax.set_title('classic_style')
plt.show()
```

## タイトル
* サブプロット、描画オブジェクトそれぞれにタイトルを指定できる

```python
fig,axes = plt.subplots(ncols=2)

# サブプロットにタイトルを指定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')

# 描画オブジェクトにタイトルを設定
fig.suptitle('figure title')

plt.show()
```

## 軸ラベル

```python
fig,ax = plt.subplots()
ax.set_xlabel('x label') # x軸にラベルを設定
ax.set_ylabel('y label') # y軸にラベルを設定

plt.show()
```

## 凡例

```python
fig, ax = plt.subplots()

# 凡例洋ラベルを設定
ax.plot([1,2,3],[2,4,9],label='legend label')
ax.legend(loc='best') # 凡例を表示

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1,2,3],[2,4,9],label='legend label')
ax.legend(loc='lower right') # 凡例の場所を動かしてみる

plt.show()
```

## ファイル出力

```python
fig, ax = plt.subplots()
ax.set_title('subplot title')
fig.savefig('sample-figure.png') # png形式で保存
fig.savefig('sample-figure.svg') # svg形式で保存
```
