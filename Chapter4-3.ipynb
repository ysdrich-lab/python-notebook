* スタイルは2種類あるｋど、この本はオブジェクト指向スタイルを使うよ

## MATLABスタイル
* 数値解析ソフトのMATLABと似た形式で描写する

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

plt.plot(x,y) # 折れ線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定

plt.show() # グラフを表示
```

## オブジェクト指向スタイル
* **描画オブジェクト**に対して**サブプロット**を追加して、サブプロットに対してグラフを描画する
* 複数のグラフをまとめて表示できるよ

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x,y) # 折れ線グラフを描画
ax.set_title('OOP-style') # タイトルを設定

plt.show()
```

# 3.グラフの種類と出力方法

## 折れ線グラフ:plot
* plotメソッドで描画する
* X座標、Y座標を表す配列、orスカラーを引数に渡す
* 点を細かく取ることで擬似的に曲線も描画できる

```python
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]

ax.plot(x, y1) # 折れ線グラフ1を描写
ax.plot(x, y2) # 折れ線グラフ2を描写

plt.show()
```

```python
import numpy as np

x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2  =np.cos(x)

fig,ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```

## 棒グラフ:bar,barh
* barメソッドで描くよ
* tick_labelでメモリに任意のラベルが指定できるよ
* bathメソッドで横向きに描けるよ
* 棒グラフの幅を指定して、自分でずらして描画すると複数の棒グラフが描けるよ

```python
fig, ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

ax.bar(x,y) # 棒グラフを描画

plt.show()
```

```python
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]
labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label = labels) # ラベルを指定

plt.show()
```

```python
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]
labels = ['spam', 'ham', 'egg']
ax.barh(x, y, tick_label = labels) # 横向きに描画

plt.show()
```

```python
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]

width = 0.4 # 棒グラフの横幅を指定
ax.bar(x, y1, width=width, tick_label = labels,label = 'y1')

# 幅をずらして2個目を描画
x2 = [num + width for num in x]
ax.bar(x2, y2, width = width, label='y2')
       
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
labels = ['spam','ham','egg']

#y1とy2を足した値を格納
y_sum = [num1 + num2 for num1,num2 in zip(y1,y2)]

#y1とy2を足した高さの棒グラフを描画
ax.bar(x, y_sum, tick_label = labels, label='y1')
ax.bar(x,y2, label='y2')
ax.legend()

plt.show()
```

## 散布図:scatter
* ランダムに生成した50個の要素を散布図で描画してみる

```python
fig, ax = plt.subplots()

# ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x,y) # 散布図を描写

plt.show()
```

* marker引数に形を指定すると描画の形が変わるよ

```python
fig,ax = plt.subplots()

# ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10],y[0:10], marker='v', label = 'triangle down') # 下向き三角
ax.scatter(x[10:20],y[10:20], marker='^', label = 'triangle up') # 上向き三角
ax.scatter(x[20:30],y[20:30], marker='s', label = 'square') # 四角
ax.scatter(x[30:40],y[30:40], marker='*', label = 'star') # 星型
ax.scatter(x[40:50],y[40:50], marker='x', label = 'x') # X

ax.legend()

plt.show()
```

## ヒストグラム:hist

```python
# データを生成
np.random.seed(123)
mu = 100 # 平均値
sigma = 15 # 標準偏差

x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x)

plt.show()
```

* 返り値として、度数分布表に使用できるデータを返す。nには各ビンの度数が格納されている
* binsにはbinの境界の値、patchesにはビンを描画するための情報が格納されている

```python
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i],bins[i+1],num))
```

* bins引数に任意の数値を指定すると、ビンの数を変更できる。
* デフォルトは10。

```python
fig, ax = plt.subplots()

ax.hist(x, bins=25) # ビンの数を指定して描画

plt.show()
```

* orientation='horizontal'で横向きに描画

```python
fig, ax = plt.subplots()

ax.hist(x, orientation='horizontal') # 横向き指定

plt.show()
```

* 棒グラフと異なり、複数の値を指定すると、自動的に横に並べて表示してくれるよ


```python
# データを生成
np.random.seed(123)
mu = 100 # 平均値
x0 = np.random.normal(mu,20,1000)

# 異なる標準偏差でデータを生成
x1 = np.random.normal(mu,15,1000)
x2 = np.random.normal(mu,10,1000)

fig, ax = plt.subplots()

labels = ['x0','x1','x2']

# 3つ分のデータのヒストグラムを描写
ax.hist((x0,x1,x2), label=labels)
ax.legend()

plt.show()
```

* 積み上げヒストグラムも描画可能

```python
fig,ax = plt.subplots()
labels = ['x0','x1','x2']

# 積み上げたヒストグラムを描画
ax.hist((x0,x1,x2), label=labels, stacked=True)
ax.legend()

plt.show()
```

## 箱ひげ図:boxplot

```python
# 異なる標準偏差でデータを生成
np.random.seed(123)
x0 = np.random.normal(0, 10, 500)
x1 = np.random.normal(0, 15, 500)
x2 = np.random.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ['x0','x1','x2']
ax.boxplot((x0,x1,x2), labels=labels) # 箱ひげ図を描画

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ['x0','x1','x2']
ax.boxplot((x0,x1,x2), labels=labels, vert=False) # 横向き箱ひげ図を描写

plt.show()
```

## 円グラフ:pie

```python
labels = ['spam','ham','egg']
x = [10,3,1]

fig, ax = plt.subplots()

ax.pie(x, labels=labels) # 円グラフを描写

plt.show()
```

* サブプロットに対して描画領域を指定する

```python
fig, ax = plt.subplots()

ax.pie(x, labels=labels)
ax.axis('equal') #アスペクト比を保持して描画する

plt.show()
```

* デフォルトでは右から反時計回りに要素を埋める
* 始める位置は、startangle
* 時計回りは、counterclockで指定

```python
fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False)
ax.axis('equal')

plt.show()
```

* 円グラフに影をつけるにはshadow
* autopctで％表記が追加される

```python
fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%') # 影と%表記を追加
ax.axis('equal')

plt.show()
```

* explode引数で一部を切り出して表示できるよ

```python
explode = [0,0.2,0] #1の要素を切り出す

fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%',
      explode=explode)
ax.axis('equal')

plt.show()
```

## 複数のグラフを組み合わせる

```python
fig, ax = plt.subplots()

x1 = [1,2,3]
y1 = [5,2,3]
x2 = [1,2,3,4]
y2 = [8,5,4,6]

ax.bar(x1, y1, label='y1') # 棒グラフを描画
ax.plot(x2,y2, label='y2') # 折れ線グラフを描画
ax.legend()

plt.show()
```

* ヒストグラムに折れ線グラフをついkあして、近似曲線を描画してみたりする

```python
np.random.seed(123)
x = np.random.randn(1000) # 正規乱数を生成

fig, ax = plt.subplots()

# ヒストグラムを描画
counts,edges, patches = ax.hist(x, bins=25)

# 近似曲線に用いる点を求める（ヒストグラムのビンの中点）
x_fit = (edges[:-1] + edges[1:])/2
# 近似曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit**2/2) / np.sqrt(2*np.pi)
ax.plot(x_fit,y)

plt.show()
```

# 4.スタイル

## 色の設定
* グラフの線とか背景色とか枠線とかの要素に色が指定できるよ。
* color引数

```python
fig, ax = plt.subplots()


ax.plot([1,3], [3,1], label='aqua', color='aqua') # 線の色を名前で指定
ax.plot([1,3],[1,3], label='#0000ff', color='#0000ff') # 16進数のRGBで指定 
ax.plot([1,3],[2,2],label='(0.1,0.2,0.5,0.3)', color=(0.1,0.2,0.5,0.3)) # RGBAをfloatで指定

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar([1],[3], color='aqua') # 塗りつぶしの色を指定
ax.bar([2],[4],color='aqua', edgecolor='black') # 塗りつぶしの色と枠線の色を指定

plt.show()
```

## 線のスタイル
* 折れ線グラフとか、グラフの枠線とか、区切り線とかのスタイル。
* 破線とか、そういうやつ。

```python
fig,ax = plt.subplots()

ax.plot([1,3],[3,1], linewidth=5.5, label='5.5') # 5.5ptの線幅で描画
ax.plot([1,3],[1,3], linewidth=10, label='10') # 10ptの線幅で描画

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1,3],[3,1],linestyle='--',label='dashed') # 破線
ax.plot([1,3],[1,3],linestyle='-.',label='dashdot') # 一点鎖線
ax.plot([1,3],[2,2],linestyle=':',label='dotted') # 点線

ax.legend()
plt.show()
```

## フォント
* タイトルとか凡例とか軸ラベルのテキストにフォントスタイルを適用できるよ

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family='fantasy', size=20,weight='bold')
ax.set_ylabel('ylabel', family='cursive', size=40, weight='light')
ax.set_title('graph title', family='monospace',size=25,weight='heavy')

plt.show()
```

```python
# フォントスタイルを辞書で定義
fontdict = {
    'family': 'fantasy',
    'size': 20,
    'weight':'normal',
}

fig,ax = plt.subplots()

#辞書形式でフォントスタイルを指定
ax.set_xlabel('xlabel',fontdict=fontdict)
ax.set_ylabel('ylabel',fontdict=fontdict)

# 個別指定したsizeで上書き
ax.set_title('graph title', fontdict=fontdict, size=40)

plt.show()
```

## テキスト描画:text

* 任意の位置にテキスト病ができるよ。フォントスタイルとかサイズも指定できるよ

```python
fig, ax = plt.subplots()

ax.text(0.2,0.4, 'Text',size=20) #テキスト描画

plt.show()
```
